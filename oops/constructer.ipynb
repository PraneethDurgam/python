```python
class Product:
    Product_id=1
    Product_name=None
    Product_price=None
    Product_qty=None

    @classmethod
    def generate_id(cls):
        id=cls.Product_id
        cls.Product_id+=1
        return id
    
    def __init__(self,name,price,qty):
        self.Product_id=self.generate_id()
        self.Product_name=name
        self.Product_price=price
        self.Product_qty=qty

    def totalprice(self):
        return self.Product_price*self.Product_qty
    
    def discount(self,coupen):
        discountprice=0
        allcoupen=[111,222,333]
        if coupen in allcoupen:
            discountprice=self.totalprice()-0.1*self.totalprice()
        else:
            discountprice=self.totalprice()
        print('--------------------------------------------')
        print('the discounted price is:',discountprice)

    def display(self):
        print('the product name:',self.Product_name)  
        print('the product price:',self.Product_price)
        print('the product qty:',self.Product_qty)
        print('the total price is:',self.totalprice())



p1=Product('marker',25,10)
p1.display()
p1.discount(111)

        
```

```python
class Grocery:
    Grocery_id = 1
    Grocery_name = None
    Grocery_category = None
    Grocery_price = None
    Grocery_qty = None

    @classmethod
    def generate_id(cls):
        id = cls.Grocery_id
        cls.Grocery_id += 1
        return id
    
    def __init__(self, name, category, price, qty):
        self.Grocery_id = self.generate_id()
        self.Grocery_name = name
        self.Grocery_category = category
        self.Grocery_price = price
        self.Grocery_qty = qty

    def totalprice(self):
        return self.Grocery_price * self.Grocery_qty

    def display(self):
        print('The grocery name:', self.Grocery_name)
        print('The grocery category:', self.Grocery_category)
        print('The grocery price:', self.Grocery_price)
        print('The grocery quantity:', self.Grocery_qty)
        print('The total price is:', self.totalprice())

g1 = Grocery('Apple', 'Fruit', 30, 10)
g2 = Grocery('Bread', 'Bakery', 25, 5)
g3 = Grocery('Milk', 'Dairy', 10, 2)


groceries = [g1, g2, g3]

for i in groceries:
    print(f'Grocery ID: {i.Grocery_id}')
    i.display()
    print('-------------------------------------------->>>>>>>>>>')
```

```python
class Sports:
    Sport_id = 1
    Sport_item = None
    Sport_type = None
    Sport_league = None

    @classmethod
    def generate_id(cls):
        id = cls.Sport_id
        cls.Sport_id += 1
        return id
    
    def __init__(self, item,type,league):
        self.Sport_id = self.generate_id()
        self.Sport_item = item
        self.Sport_type = type
        self.Sport_league = league

    def display(self):
        print('The sport item:', self.Sport_item)
        print('The sport type:', self.Sport_type)
        print('The sports league:', self.Sport_league)
       
s1 = Sports('Soccer Ball', 'Soccer', "FIFA",)
s2 = Sports('Basketball', 'Basketball','NBA' )
s3 = Sports('Tennis Racket', 'Tennis', 'WIMBELDON')
s4 = Sports('Cricket Bat', 'Cricket', 'ICC WORLD CUP')

sports_items = [s1, s2, s3, s4]

for s in sports_items:
    print(f'Sport ID: {s.Sport_id}')
    s.display()
    print('--------------------------------------------')
```

```python
class Cloth:
    Cloth_id = 1
    Cloth_type = None
    Cloth_size = None
    Cloth_price = None
    Cloth_qty = None

    @classmethod
    def generate_id(cls):
        id = cls.Cloth_id
        cls.Cloth_id += 1
        return id
    
    def __init__(self,type, size, price, qty):
        self.Cloth_id = self.generate_id()
        self.Cloth_type = type
        self.Cloth_size = size
        self.Cloth_price = price
        self.Cloth_qty = qty

    def totalprice(self):
        return self.Cloth_price * self.Cloth_qty
    
    def discount(self, coupon):
        discountprice = 0
        allcoupons = ['abc','xyz']
        if coupon in allcoupons:
            discountprice = self.totalprice() - 0.2 * self.totalprice()
        else:
            discountprice = self.totalprice()
        print('--------------------------------------------')
        print('The discounted price is:', discountprice)

    def display(self):
        print('The cloth type:', self.Cloth_type)
        print('The cloth size:', self.Cloth_size)
        print('The cloth price:', self.Cloth_price)
        print('The cloth quantity:', self.Cloth_qty)
        print('The total price is:', self.totalprice())


c1 = Cloth('T-shirt', 'M', 500, 3)
c1.display()
c1.discount('abc')
```

```python
class Employee:
    ename=None
    eid=None

    def __init__(self,ename,eid):
        self.ename=ename
        self.eid=eid
    
class Employee1(Employee):
    esal=None

    def __init__(self, ename, eid,esal):
        super().__init__(ename, eid)
        self.esal=esal
    
    def display(self):
        print('employee name',self.ename)
        print('employee id ',self.eid)
        print('employee sal ',self.esal)

e1=Employee1('king',25,5000)
e1.display()
```

```python
class Hospital:
    name=None
    address=None
    phone=None
    def __init__(self,name,address,phone):
        self.name=name
        self.address=address
        self.phone=phone
    
    def displayhospital(self):
        print('hosiptal name:',self.name)
        print('hosiptal adress:',self.address)
        print('hosiptal phone:',self.phone)

class Doctor(Hospital):
    Doctor_name=None
    Doctor_phone=None

    def __init__(self,name,address,phone,Doctor_name,Doctor_phone):
        Hospital.__init__(self,name,address,phone)
        self.Doctor_name=Doctor_name
        self.Doctor_phone=Doctor_phone

    def displaydoctor(self):
        super().displayhospital()
        print('------------------------------')
        print('doctor name:',self.Doctor_name)
        print('doctor phone:',self.Doctor_phone)
        
h1=Doctor('mallareddy','hyd',123456789,'king',25285)
h1.displaydoctor()
```
